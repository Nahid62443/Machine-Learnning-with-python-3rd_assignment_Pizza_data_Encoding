# access in one technic another way

```python
df=pd.read_csv('Pizza Data.csv')
df.head()
```

```python
columns=['Size']
for column in columns:
    one=pd.get_dummies(df[column],drop_first=True)
    df=pd.concat((df,one),axis=1).drop(column,axis=1)
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
df=pd.read_csv('Pizza Data.csv')
```

```python
df.head()
```

```python
df['Size'].unique()
```

```python
All_over_pizza_size=['Small (10")', 'Medium (12")', 'Large (14")', 'X-Large (16")',
       'Medium', 'Large', 'Personal', 'Small', 'Mini', 'Jumbo',
       'X Large (16")']
```

```python
ordinal=OrdinalEncoder(categories=[All_over_pizza_size])
encoded=ordinal.fit_transform(df[['Size']])
```

```python
encoded
```

```python
encoded_size=pd.DataFrame(encoded,columns=["Encoded"])
```

```python
encoded_size
```

```python
df.Size=encoded_size
```

```python
df.Size
```

# Access in one technic

```python
df=pd.read_csv("Pizza Data.csv")
df.head()
```

```python
columns=['Size']
for col in columns:
    uniques=df[col].unique()
    df[col]=OrdinalEncoder(categories=[uniques ]).fit_transform(df[[col]])
```
