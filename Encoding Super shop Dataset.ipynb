# Label Encoeder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_obj=LabelEncoder()
```

```python
ds.Area=label_obj.fit_transform(ds["Area"])
```

```python
ds.Area
```

```python
ds.Area.head()
```

```python
for column in ds.columns:
    if ds[column].dtype ==np.number:
        continue
        ds[column]=label_obj.fit_transform(ds[column])
```

```python
ds.head()
```

```python
#another way
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in ds.columns:
    if is_numeric_dtype(ds[column]):
        continue
    else:
        ds[column]=LabelEncoder().fit_transform([column])
```

```python
ds.head()
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
df=pd.read_csv('supershop.csv')
df.head()
```

```python
df.Area.unique()
```

```python
city=["Dhaka","Ctg","Rangpur"]
```

```python
ordinal=OrdinalEncoder(categories=[city])
encoded=ordinal.fit_transform(df[['Area']])
```

```python
encoded
```

```python
encoded_area=pd.DataFrame(encoded,columns=["Encoded"])
```

```python
encoded_area
```

```python
df.Area=encoded_area
```

```python
df.Area
```

```python
df=pd.read_csv('supershop.csv')
df.head()
```

```python
columns=['Area']
for col in columns:
    uniques=df[col].unique()
    df[col]=OrdinalEncoder(categories=[uniques ]).fit_transform(df[[col]])
```
